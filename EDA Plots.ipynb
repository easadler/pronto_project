```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
from scipy.stats import gaussian_kde

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
%matplotlib inline
```

```python
df = pd.read_csv('new_trips.csv')
```

```python
df.columns
```

```python
df_days = df.groupby(['dayofweek'])['bikeid'].count().reset_index()
df_days.plot(kind='bar')
```

```python
df_hours = df.groupby(['starthour'])['bikeid'].count().reset_index()
df_hours.plot(kind='bar')
```

```python
df_hours = df.groupby(['hood'])['bikeid'].count().reset_index()
df_hours.plot('hood','bikeid',kind='bar')
```

```python
df_hours = df.groupby(['dest_hood'])['bikeid'].count().reset_index()
df_hours.plot('dest_hood','bikeid',kind='bar')
```

```python
df_hours = df.groupby(['month'])['bikeid'].count().reset_index()
df_hours.plot('month','bikeid',kind='bar')
```

```python
df_hours = df.groupby(['month','usertype'])['bikeid'].count().unstack().reset_index()
df_hours.plot('month', kind='bar', figsize = (10,10))
```

```python
df_hours = df.groupby(['hood','dayofweek'])['bikeid'].count().unstack().reset_index()
df_hours.plot('hood', kind='bar', figsize = (40,15))
```

```python
df_hours = df.groupby(['dayofweek','usertype'])['bikeid'].count().unstack().reset_index()
df_hours.plot('dayofweek', kind='bar', figsize = (10,10))
```

```python
df_hours = df.groupby(['season','usertype'])['bikeid'].count().unstack().reset_index()
df_hours.plot('season', kind='bar', figsize = (10,10))
```

```python
df_hours = df.groupby(['season','dayofweek'])['bikeid'].count().unstack().reset_index()
df_hours.plot('season', kind='bar', figsize = (10,10))
```

```python
df_hours = df.groupby(['hood','season'])['bikeid'].count().unstack().reset_index()
df_hours.plot('hood', kind='bar', stacked=True, figsize = (10,10))
```

```python
df_hours = df.groupby(['hood','dest_hood'])['bikeid'].count().unstack().reset_index()
df_hours.plot('hood', kind='bar',stacked=True, figsize = (30,30))
```

```python
df_hours = df.groupby(['dest_hood','hood'])['bikeid'].count().unstack().reset_index()
df_hours.plot('dest_hood', kind='bar',stacked=True, figsize = (30,30))
```
