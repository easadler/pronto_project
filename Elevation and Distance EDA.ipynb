```python
from pymongo import MongoClient
import requests
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier

%matplotlib inline
```

```python
df_stations = pd.read_csv('open_data_year_one/2015_station_data.csv')
```
